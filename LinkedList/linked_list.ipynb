<h1>LINKED LIST</h1> (Linear Data Sturcture - Collection of Nodes(Data + Address To Next Node))

ADVANTAGES:
1. Dynamic DS that can be grown or shrinked in size at runtime.
2. Insertion & Deletion are inexpensive operations in LL.
3. No need for contiguous storage on memory heap.

DISADVANTAGES:
1. Accessing an element is costly compared to arrays as we cannot access any index in the LL directly.

Types of LL:
1. Singly LL
2. Doubly LL
3. Circular LL
4. Circular Doubly LL

INSERSTION & DELETION IN A SINGLY LINKED LIST

```python
class Node:
    def __init__(self,data, next_node = None) -> None:
        self.data = data
        self.next_node = next_node
    
    def __str__(self) -> str:
        """this magic method is automatically called when we pass the object into a
        print statement and the value returned by the same is printed"""
        return str({'data' : self.data, 'next_node': self.next_node})
    
class LinkedList:
    def __init__(self, head = None) -> None:
        self.head = head

    #INSERTION IN A LINKED LIST
    def insert_at_head(self,node):
        'inserts a at head of the linked list and returns true if the node was successfully inserted'
        try:
            if self.head != None:
                node.next_node = self.head
                self.head = node
            else:
                self.head = node
            return True
        except:
            return False
        
    def insert_at_tail(self,node):
        'inserts a node at the tail of the linked list and returns true if the node was successfully inserted'
        try:
            temp = self.head
            if temp != None:
                while temp.next_node != None:
                    temp = temp.next_node
                temp.next_node = node
            else:
                temp = node
            return True
        except:
            return False
        
    def insert_in_middle(self,node,index):
        'inserts a node at the input position of the linked list and returns true if the node was successfully inserted'
        try:
            if index == 0:
                self.insert_at_head(node)
            elif index == self.length_of_ll():
                self.insert_at_tail(node)
            else:
                temp = self.head
                counter = 0
                while temp != None  and index - 1 > counter:
                    temp = temp.next_node
                    counter += 1
                node.next_node = temp.next_node
                temp.next_node = node
            return True
        except:
            return False
        
    #DELETION IN A LINKED LIST (Note :- Logic to free up the memory after deletion needs to be considered not implemented here.)
    def delete_by_data(self,data):
        try:
            temp = self.head
            previous_node = None
            while temp != Node:
                if temp.data == data:
                    if previous_node != None:
                        previous_node.next_node = temp.next_node
                    else:
                        self.head = temp.next_node
                else:
                    previous_node = temp
                temp = temp.next_node
            return True
        except:
            return False
    
    def delete_by_position(self,index):
        try:
            if index == 0:
                self.head = self.head.next_node
            else:
                temp = self.head
                current_index = 0
                while temp != None and current_index < index - 1:
                    current_index += 1
                    temp = temp.next_node
                temp.next_node = temp.next_node.next_node
            return True
        except:
            return False
        
    #UTILITY FUNCTIONS LINKED LIST
    def length_of_ll(self):
        'return the length of the linked list'
        length = 0
        temp = self.head
        while temp != None:
            temp = temp.next_node
            length += 1
        return length
    
    def __str__(self) -> str:
        temp = self.head
        linked_list = ''
        while temp != None:
            if temp.next_node != None:
                linked_list = linked_list + str(temp.data) + '->'
            else:
                linked_list = linked_list + str(temp.data)
            temp = temp.next_node

        return linked_list 
```

REVERSE NODES IN K GROUPS

```python
# Definition for singly-linked list.
# class ListNode(object):
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution(object):
    def length_of_ll(self, head):
        'return the length of the linked list'
        length = 0
        temp = head
        while temp != None:
            temp = temp.next
            length += 1
        return length

    def reverseList(self, head, k):
        """
        :type head: ListNode
        :rtype: ListNode
        """
        previous_node = None
        elements_iterated = 0
        tail = head
        while head != None and elements_iterated < k:
            current_node = head
            head = head.next
            current_node.next = previous_node
            previous_node = current_node
            elements_iterated += 1
        current_node = head
        head = previous_node
        return head,tail,current_node

    def reverseKGroup(self, head, k):
        """
        :type head: ListNode
        :type k: int
        :rtype: ListNode
        """
        if head != None:
            head,tail,current_node = self.reverseList(head, k)
            if self.length_of_ll(current_node) >= k:
                tail.next = self.reverseKGroup(current_node,k)
            else:
                tail.next = current_node
            return head
        else:
            return None
```

CYCLE DETECTECTION IN A LINKED LIST
1. Solution 1: Use a set and keep a tab whether a new node occuring has occured previously as well if yes in that case there is a cycle present in the linked list.
2. Solution 2: <strong>Floyd's cycle detection algorithm</strong> maintain two reference variables and update one of them with one step at a time and the other one with 2 steps at a time if at any given point both of these become equal that we can say there exists a cycle in the linked list.
<br><strong>Note : if we want to find the position where that loop exists we need to start a reference variable from head of linked list and other reference variable from the point where slow and fast pointers became equal and then update them till the time the both become equal and that is the point where the loop exists.</strong>

```python
# Definition for singly-linked list.
# class ListNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution(object):
    def hasCycle_soultion1(self, head):
        """
        :type head: ListNode
        :rtype: bool
        """

        #Solution 1 Using Set
        node_data = set()
        cycle_present = False
        temp = head
        while temp != None:
            if temp.next in node_data:
                cycle_present = True
                break
            node_data.add(temp)
            temp = temp.next
        return cycle_present
    
    def hasCycle_solution2(self,head):

        #Solution 2 Using Floyd's Cycle Detection Algorithm
        slow = head
        fast = head
        cycle_present = False

        while fast != None:
            if fast.next != None:
                fast = fast.next.next
            else:
                fast = fast.next
            slow = slow.next
            if fast == slow and fast != None and slow != None:
                cycle_present = True
                break
        return cycle_present
            
```

MERGING TWO SORTED LINKED LIST

```python
# Definition for singly-linked list.
# class ListNode(object):
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution(object):
    def mergeTwoLists(self, list1, list2):
        """
        :type list1: Optional[ListNode]
        :type list2: Optional[ListNode]
        :rtype: Optional[ListNode]
        """
        #Initializing the head node for the final linked list
        final_head = None
        if list1 != None and list2 != None:
            if list1.val <= list2.val:
                final_head = list1
                list1 = list1.next
            else:
                final_head = list2
                list2 = list2.next
        elif list1 != None and list2 == None:
            return list1
        elif list2 != None and list1 == None:
            return list2
        
        temp = final_head

        #Keep merging till we dont reach end of one of the input linked list
        while list1 != None and list2 != None:
            if list1.val <= list2.val:
                temp.next = list1
                list1 = list1.next
            elif list1.val > list2.val:
                temp.next = list2
                list2 = list2.next
            temp = temp.next

        #Merge to the final linked list if one of the two linked list is left to be merged
        if list1 == None and list2 != None:
            temp.next = list2
        elif list2 == None and list1 != None:
            temp.next = list1
        return final_head
```

FINDING MIDDLE OF A LINKED LIST

```python
# Definition for singly-linked list.
# class ListNode(object):
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution(object):
    def middleNode(self, head):
        """
        :type head: ListNode
        :rtype: ListNode
        """
        #make one pointer run at twice the speed of the other one
        slow_pointer = head
        fast_pointer = head

        while fast_pointer != None:
            if fast_pointer.next != None:
                fast_pointer = fast_pointer.next.next
                slow_pointer = slow_pointer.next
            else:
                fast_pointer = fast_pointer.next
        return slow_pointer
```

```python
# Definition for singly-linked list.
# class ListNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution(object):
    def detectCycle(self, head):
        """
        :type head: ListNode
        :rtype: ListNode
        """
        slow = head
        fast = head
        temp = None
        while fast != None and fast.next != None:
            slow = slow.next
            fast = fast.next.next
            if slow == fast:
                temp = slow
                break

        if temp != None:
            temp_head = head
            while temp_head != None:
                if temp_head == temp:
                    return temp_head
                else:
                    temp_head = temp_head.next
                    temp = temp.next
        else:
            return None
```

REMOVE NODES FROM LINKED LIST (Remove every node which has a node with a strictly greater value anywhere to the right side of it.)

```python
# Definition for singly-linked list.
# class ListNode(object):
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution(object):
    def removeNodes(self, head):
        """
        :type head: Optional[ListNode]
        :rtype: Optional[ListNode]
        """

        if head == None:
            return None
        else:
            child_head = self.removeNodes(head.next)
            if child_head != None and child_head.val > head.val:
                head = child_head
                return head
            else:
                head.next = child_head
                return head
```

```python
class Node:
    def __init__(self,data,next_node = None)-> None:
        self.data = data
        self.next_node = next_node
class CircularLinkedList:
    def __init__(self,tail = None) -> None:
        self.tail = tail #in case of CLL we can reach the head if we know the tail of the list.So maintaining a tail in a CLL is mandatory.

    def insert_node(self,element,node):
        #assuming that the element that after which we wish to do the insertion is present in the circular linked list.
        if self.tail == None:
            self.tail = node
            node.next_node = self.tail
        else:
            temp = self.tail.next_node
            reached_end = True
            while temp != self.tail:
                if temp.data == element:
                    current = temp
                    reached_end = False
                    break
                temp = temp.next_node
            if reached_end:
                current = self.tail
            node.next_node = current.next_node
            current.next_node = node
            if reached_end:
                self.tail = node
    
    def delete_node(self, element):
        if self.tail == None:
            print("No such element exists in the circular linked list.")
            return
        element_found = False
        temp = self.tail.next_node
        previous = self.tail
        while temp != self.tail:
            if temp.data == element:
                previous.next_node = temp.next_node
                temp.next_node = None
                element_found = True
                break
            previous = temp
            temp = temp.next_node
        if temp == self.tail and temp.data == element:
            previous.next_node = temp.next_node
            temp.next_node = None
            element_found = True
            self.tail = previous
        if not element_found:
            print("No such element exists in the circular linked list.")

    #UTILITY FUNCTIONS
    def __str__(self) -> str:
        if self.tail == None:
            return "Empty Circular Linked List"
        temp = self.tail.next_node
        cll = ''
        while temp != self.tail:
            cll = cll + str(temp.data) + '->'
            temp = temp.next_node
        cll = cll + str(temp.data)
        return cll
        
```
